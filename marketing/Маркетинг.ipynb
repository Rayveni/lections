### Увеличение размера покупки
* **Upsales**
Предлагаем улучшенную ,более дорогую версию продукта.<br>Например,при заселении в отель, Вам предлагают улучшить комнату всего за 6$
И по сути — что такое 6 долларов, когда вы уже заплатили 100-300 долларов за ночь? <br>Таким образом, решаются две задачи: **максимизация прибыли и более эффективное управление инветарём**.
* **Cross sales** Продаём продукты, которые могут каким-либо образом дополнять ваше основное предложение.
* **Sets** по сути те же Cross sales , но как правило содержит в себе товары, по отдельности стоящие дороже

### Увеличение периодичности покупок
* используйте новости, связанные с компанией
Дополнительным инфоповодом и тригером для новой покупки является ситуация, которая происходит с компанией. К различным номинациям и победам в конкурсах (особенно если таковых не имеется) можно смело добавить всевозможные «Празднуем 5 лет на рынке» и прочие дни рождения директоров или хорошее настроение секретарши Тамары. 
* ситуативные акции/скидки
Например по промо коду в честь воскресенья, чтобы получить минус 5 долларов от покупки, или другому промокоду для тех, кто проснулся только к понедельнику, но уже за 3-долларовую скидку.(борьба с негативом)
