# Значимость корреляции
## Критерий Стьюдента для корреляции Пирсона и Спирмена

выборки: $(X_{1i},X_{2i}),i=1,...,n$<br>
нулевая гипотеза: $H_0:r_{X_1X_2}=0$<br>
альтернатива: $H_1:r_{X_1X_2}<\not =>0$<br>
статистика: $T=\frac{r_{X_1X_2}\sqrt{n-2}}{\sqrt{1-r_{X_1X_2}^2}}$<br>
нулевое распределение: $T\sim St(n-2)$

## Критерий $\chi^2$ (для корреляции Мэтьюса)
выборки: $(X_{1i},X_{2i}),i=1,...,n$<br>
$X_1,X_2 \in \{0,1\}$<br>
нулевая гипотеза: $H_0:MCC_{X_1X_2}=0$<br>
альтернатива: $H_1:MCC_{X_1X_2}\not=0$<br>
статистика: $\chi^2=n MCC_{X_1X_2}^2$<br>
нулевое распределение: $\chi^2 \sim \chi^2_1$

**условия применимости:**<br>
* $n \geq 40$
* $\frac{(a+c)(a+b)}{n}>5,\frac{(a+c)(c+d)}{n}>5,\frac{(b+d)(a+b)}{n}>5,\frac{(b+d)(c+d)}{n}>5$


## Критерий $\chi^2$ (для категориальных признаков)

  $X_1\X_2$ | 1 | .. | j| .. | K_2 |$\sum$
  ------------- | -------------|
  **1**  |  |  | |  | |
  **...** |  |  | |  | |
  i |  |  | $n_{ij}$|  | |$n_{i+}$
**...** |  |  | |  | |
**K_1** |  |  | |  | |
**$\sum$** |  |  | $n_{j+}$|  | |

выборки: $(X_{1i},X_{2i}),i=1,...,n$<br>
$X_1 \in \{1,..,K_1\},X_2 \in \{1,..,K_2\}$<br>
нулевая гипотеза: $H_0:X_1$ и $X_2$ независимы<br>
альтернатива: $H_0$ неверна<br>
статистика: $$\chi^2(X_1^n,X_2^n)=\sum^{K_1}_{i=1}\sum^{K_2}_{j=1} \frac{\Big (n_{ij}-\frac{n_{i+}n_{j+}}{n}\Big )^2}{\frac{n_{i+}n_{j+}}{n}}=n \Bigg ( \sum^{K_1}_{i=1}\sum^{K_2}_{j=1} \frac{n_{ij}^2}{n_{i+}n_{j+}}-1  \Bigg )$$
нулевое распределение: $\chi^2(X_1^n,X_2^n) \sim \chi ^2_{(K_1-1)(K_2-1)}$

**условия применимости:**<br>
* $n \geq 40$
* $\frac{n_{i+}n_{j+}}{n}<5$ не более чем в 20% ячеек

#### Коэффициент Крамера

Можно считать, что для категориальных признаков критерий хи-квадрат проверяет значимость отличия от нуля коэффициента V Крамера против альтернативы что он нулю не равен:
$$\phi_c(X_1^n,X_2^n)=\sqrt{\frac{\chi^2(X_1^n,X_2^n)}{n(min(K_1,K_2)-1)}}$$
